```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sb
import numpy as np
```

## How the base plane is created

```python
ax = plt.axes(projection="3d")

x_data = np.arange(0,50,0.1)
y_data = np.arange(0,50,0.1)

x_data = np.array([0, 10, 10, 0])
y_data = np.array([0, 0, 10, 10])
Z = np.array([[0,0,0,0]])

X, Y = np.meshgrid(x_data, y_data)

ax.plot_surface(X, Y, Z)

plt.xlabel('X label')
plt.ylabel('Y label')

plt.show()
```
